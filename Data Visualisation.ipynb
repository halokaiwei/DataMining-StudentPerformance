```python
%pip install seaborn
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv("StudentPerformance_FinalVersion5.csv")
plt.figure(figsize=(10,6))
data['binned_average_marks'] = pd.cut(data['average_marks'], bins=range(50, 90, 5))

data['parental level of education'] = pd.Categorical (data['parental level of education'],
                                                      categories=["some high school",
                                                                  "high school",
                                                                  "some college",
                                                                  "associate's degree",
                                                                  "bachelor's degree",
                                                                  "master's degree"],
                                                      ordered=True)
sns.countplot(x='parental level of education', data=data, hue='binned_average_marks')
plt.title('Parental Education Level and Average Marks Group Analysis')
plt.show()


```

```python
%pip install seaborn
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import ticker
import seaborn as sns
data= pd.read_csv("StudentPerformance_FinalVersion5.csv")

bins = [0, 55, 60, 65, 70, 75, 80, 85]
bin_labels = ['50-55','55-60','60-65','65-70','70-75','75-80','80-85']

data['average_marks_range'] = pd.cut(data['average_marks'], bins=bins, labels=bin_labels)

ax = sns.countplot(x='average_marks_range', hue='test preparation course', data=data)
ax.yaxis.set_major_locator(ticker.MultipleLocator(10))

plt.xlabel('Average Marks Range')
plt.ylabel('Count')
plt.title('Relationship between Test Preparation and Average Marks Range')
plt.show()


```
